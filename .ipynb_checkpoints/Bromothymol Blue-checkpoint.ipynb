```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import interpolate
import pandas as pd
from numpy import trapz

%matplotlib inline
```

```python
# The following are spectra of each bromothymol blue conjugate. The data was digitized from open source lit
BTB_Acid = pd.read_csv('BTB_Acid.csv')
BTB_Base = pd.read_csv('BTB_Base.csv')

# The following are the normalized peak absorption for each conjugate as a function of pH.  The data was digitized from open source lit
pka_data_blu = pd.read_csv('Blu_Form.csv')
pka_data_yel = pd.read_csv('Yel_Form.csv')
```

```python
# As the pka data is normalized, we need to calculate the ratio of the spectra so that we can model the pka's relative impact.
ratio = np.amax(BTB_Acid['abs'])/np.amax(BTB_Base['abs'])
```

```python
# We are trimming the data
wave = np.arange(400, 704, 5)

f = interpolate.interp1d(BTB_Acid['wave'], BTB_Acid['abs'])
BTB_Acid = f(wave)

g = interpolate.interp1d(BTB_Base['wave'], BTB_Base['abs'])
BTB_Base = g(wave)

pH = np.arange(50, 91, 1)/10

f = interpolate.interp1d(pka_data_yel['ph'], pka_data_yel['abs'])
pka_yellow = f(pH)

g = interpolate.interp1d(pka_data_blu['ph'], pka_data_blu['abs'])
pka_blu = g(pH)
```

```python
# Place pertinent data in a dataframe

df = pd.DataFrame(list(zip(pH, pka_blu, pka_yellow)), columns =['pH', 'Base', 'Acid'])
df.head()
```

```python
# Plot the results; the bold lines are within the physiological pH range
for i in df['pH'][::1]:
    
    if i <= 7.4 and i >= 6.8:
        alpha = 1
        plt.plot(wave, [x*df.loc[df['pH'] == i].Base.values + y*df.loc[df['pH'] == i].Acid.values for x, y in zip(BTB_Base/np.amax(BTB_Base), BTB_Acid/np.amax(BTB_Base))], alpha=alpha, label ='pH =' +str(i))
    else:
        alpha = .2
        plt.plot(wave, [x*df.loc[df['pH'] == i].Base.values + y*df.loc[df['pH'] == i].Acid.values for x, y in zip(BTB_Base/np.amax(BTB_Base), BTB_Acid/np.amax(BTB_Base))], alpha=alpha)
  
plt.grid()
plt.legend()
plt.ylabel('Abs\n', fontsize = 14)
plt.xlabel('\nWavelength (nm)', fontsize = 14)
plt.title('Bromothymol Blue\n', fontsize = 16)
plt.savefig('demo.png', transparent=True)
plt.show()
```

```python
# Isolate the spectra within the physiological pH range of interest
spectra68 = np.array([x*df.loc[df['pH'] == 6.8].Base.values + y*df.loc[df['pH'] == 6.8].Acid.values for x, y in zip(BTB_Base/np.amax(BTB_Base), BTB_Acid/np.amax(BTB_Base))]).flatten()
spectra69 = np.array([x*df.loc[df['pH'] == 6.9].Base.values + y*df.loc[df['pH'] == 6.9].Acid.values for x, y in zip(BTB_Base/np.amax(BTB_Base), BTB_Acid/np.amax(BTB_Base))]).flatten()
spectra70 = np.array([x*df.loc[df['pH'] == 7.0].Base.values + y*df.loc[df['pH'] == 7.0].Acid.values for x, y in zip(BTB_Base/np.amax(BTB_Base), BTB_Acid/np.amax(BTB_Base))]).flatten()
spectra71 = np.array([x*df.loc[df['pH'] == 7.1].Base.values + y*df.loc[df['pH'] == 7.1].Acid.values for x, y in zip(BTB_Base/np.amax(BTB_Base), BTB_Acid/np.amax(BTB_Base))]).flatten()
spectra72 = np.array([x*df.loc[df['pH'] == 7.2].Base.values + y*df.loc[df['pH'] == 7.2].Acid.values for x, y in zip(BTB_Base/np.amax(BTB_Base), BTB_Acid/np.amax(BTB_Base))]).flatten()
spectra74 = np.array([x*df.loc[df['pH'] == 7.4].Base.values + y*df.loc[df['pH'] == 7.4].Acid.values for x, y in zip(BTB_Base/np.amax(BTB_Base), BTB_Acid/np.amax(BTB_Base))]).flatten()
```

```python
# Convert to transmission (the transmission spectra will be used later on in the model)
plt.plot(wave, 10**(-spectra68), label = 'pH = 6.8')
plt.plot(wave, 10**(-spectra69), label = 'pH = 6.9')
plt.plot(wave, 10**(-spectra70), label = 'pH = 7.0')
plt.plot(wave, 10**(-spectra71), label = 'pH = 7.1')
plt.plot(wave, 10**(-spectra72), label = 'pH = 7.2')
plt.plot(wave, 10**(-spectra74), label = 'pH = 7.4')
plt.ylabel('Transmission \n', fontsize = 14)
plt.xlabel('\nWavelength (nm)', fontsize = 14)
plt.title('Bromothymol Blue \n', fontsize=16)
plt.grid()
plt.legend()
plt.show()
```
